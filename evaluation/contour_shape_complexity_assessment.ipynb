```python
def _compress(gt: np.ndarray) -> bytes:
    data = np.packbits(gt)
    data = bz2.compress(data=data, compresslevel=9)
    return data, gt.shape

def _decompress(data: bytes, shp: tuple) -> np.ndarray:
    data = bz2.decompress(data=data)
    data = np.frombuffer(data, dtype=np.uint8)
    gt = np.unpackbits(data)
    gt = gt[:np.prod(shp)].reshape(shp)
    return gt > 0
```

**Minkowski–Bouligand Dimension**, Sources: [stackoverflow.com](https://stackoverflow.com/questions/44793221/python-fractal-box-count-fractal-dimension), [wikipedia.org](https://en.wikipedia.org/wiki/Minkowski%E2%80%93Bouligand_dimension)

```python
def fractal_dimension(gt):
    assert gt.ndim == 2 and str(gt.dtype) == 'bool'
    def boxcount(gt, k):  # From https://github.com/rougier/numpy-100 (#87)
        s = np.add.reduceat(
            np.add.reduceat(gt, np.arange(0, gt.shape[0], k), axis=0),
            np.arange(0, gt.shape[1], k), axis=1)
        return len(np.where((s > 0) & (s < k ** 2))[0])  # Count non-empty (0) and non-full boxes (k ** 2)

    p = np.amin(gt.shape)  # Minimum dimension of image
    n = 2 ** np.floor(np.log(p) / np.log(2))  # Greatest power of 2 less than or equal to p
    n = int(np.log(n) / np.log(2))  # Extract the exponent
    sizes = 2 ** np.arange(n, 1, -1)  # Build successive box sizes (from 2 ** n down to 2 ** 1)
    counts = [boxcount(gt, size) for size in sizes]  # Actual box counting with decreasing size
    coeffs = np.polyfit(np.log(sizes), np.log(counts), 1)  # Fit the successive log(sizes) with log(counts)
    return -coeffs[0]
```

```python
def print_dims(gts, gt_ids):
    hr_2_gt_ids = dict(map(reversed, gt_ids_2_human_readable.items()))
    res_mbd = sorted((fractal_dimension(gt), gt_ids_2_human_readable[gt_id]) for gt, gt_id in zip(gts, gt_ids))
    for mbd, idh in res_mbd:
        print(f'Minkowski–Bouligand dimension: {mbd:0.2f}, {100 * mbd / res_mbd[-1][0]:0.1f}% {idh}',
              f'\t(compressed size: {len(data[hr_2_gt_ids[idh]][0])} bytes)')
```
